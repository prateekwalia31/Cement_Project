## Removing Outliers.

```python
def remove_outliers(data, col_names, n_std_dev):

  ''' Removing outliers from the dataframe columns '''

  clean_df = df.copy()

  for column in col_names:

    column_mean = clean_df[column].mean()
    column_std = clean_df[column].std()
    threshold = column_mean + n_std_dev*column_std
    clean_df.loc[clean_df[column] > threshold, column] = pd.NA

  
  return clean_df
```

```python
if 'Age' in column_names:
  column_names.remove('Age')

new_df = remove_outliers(df, column_names, 3)

new_df.isna().sum()
```

# Data Splitting

```python
from sklearn.model_selection import train_test_split

y = normalized_df['CompressiveStrength'].copy()

X = normalized_df.drop(['CompressiveStrength'], axis=1).copy()

X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.70, random_state=15)  # 70% For Training and 30% for Testing and Validation

X_val, X_test, y_val, y_test = train_test_split(X_rem,y_rem, test_size=0.67, random_state=15)

X_concat = pd.concat([X_train, X_val])

y_concat = pd.concat([y_train, y_val])
```

# Model Development and Evaluation (I)

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

## Model Evaluation Function

```python
metric_dict = {}

def regressor_evaluation(actual_values, predicted_values,model_name,metric_dict):

  ''' Function that takes the model predicted values and evaluates the four metrics using testing and training data.'''


  rmse = sqrt(mean_squared_error(actual_values,predicted_values))
  mse = mean_squared_error(actual_values,predicted_values)
  mae = mean_absolute_error(actual_values,predicted_values)
  r2 = r2_score(actual_values,predicted_values)

  print('R squared score:'  ,round(r2,4),'\n')
  print('Root Mean Squared error:', round(rmse,4),'\n')
  print('Mean Squared error:' ,round(mse,4),'\n')
  print('Mean Absolute Error:', round(mae,4),'\n')

  metric_dict[model_name] = {'R-squared': round(r2, 4), 'RMSE': round(rmse, 4), 
                              'MSE': round(mse, 4), 'MAE': round(mae, 4)}


```

## KNN Regression Model

```python
# KNN Regression Model

from sklearn.neighbors import KNeighborsRegressor as KNNr
from sklearn.model_selection import GridSearchCV


params = {'n_neighbors': [3,4,5,6,7,8,9,10],
              'weights': ['uniform', 'distance'],
              'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}


model_knn_regressor = KNNr()

grid = GridSearchCV(estimator=model_knn_regressor, param_grid = params, cv=4)

grid.fit(X_concat, y_concat)

print("Best hyperparameters:", grid.best_params_)

best_knn_model = grid.best_estimator_


knn_train_pred = best_knn_model.predict(X_concat)
knn_test_pred = best_knn_model.predict(X_test)


print("\nKNN Evaluation Report\n")
print('--Testing Evaluation--\n')
regressor_evaluation(y_test,knn_test_pred,'KNN', metric_dict)

print('\n--Training Evaluation--\n')
regressor_evaluation(y_concat,knn_train_pred,'KNN_training',metric_dict)

```

### Model Evaluation Visualization Function

```python
# Metric Evaluation visualization function definition.

def eval_visual(i,j,k,l):

  '''Function takes in the name of the model as key for metric_dict dictionary and creates a bar chart for comparison for the four metrics. '''

  r2_values_ml = [metric_dict[i]['R-squared'],metric_dict[j]['R-squared'],metric_dict[k]['R-squared'],metric_dict[l]['R-squared']]
  rmse_values_ml = [metric_dict[i]['RMSE'],metric_dict[j]['RMSE'],metric_dict[k]['RMSE'],metric_dict[l]['RMSE']]
  mse_values_ml = [metric_dict[i]['MSE'],metric_dict[j]['MSE'],metric_dict[k]['MSE'],metric_dict[l]['MSE']]
  mae_values_ml = [metric_dict[i]['MAE'],metric_dict[j]['MAE'],metric_dict[k]['MAE'],metric_dict[l]['MAE']]

  print('\n\n')

  fig, ax = plt.subplots(nrows=1, ncols = 4, figsize=(28,12))

  colors = ['r','m','b','g']

  ax[0].bar([i, j, k, l], r2_values_ml, color=colors)
  ax[0].set_title('R-Squared (Higher is better)')

  ax[1].bar([i, j, k, l], rmse_values_ml,color=colors)
  ax[1].set_title('RMSE (Lower is better)')
  
  ax[2].bar([i, j, k, l], mse_values_ml, color=colors)
  ax[2].set_title('MSE (Lower is better)')

  ax[3].bar([i, j, k, l], mae_values_ml,color=colors)
  ax[3].set_title('MAE (Lower is better)')

  plt.show()
  print('\n\n')
```

## Model Evaluation using Elbow mehod and Silhouette Scores

```python
def best_model_elbow(data, k_max):

  sqe = []

  for k in range(2, k_max+1):

    kmeans_model = KMeans(n_clusters = k, n_init='auto', random_state=15)

    kmeans_model.fit(data)

    sqe.append(kmeans_model.inertia_)

  plt.plot(range(2, k_max+1), sqe)
  plt.title("Sum of Squared Distance vs K")
  plt.xlabel('Number of Clusters (K)')
  plt.ylabel('Sum of Squared Distances')

  plt.show()

  print('\n\n')

  #print(sqe)

  sil_scores = []

  for k in range(2, k_max+1):
    kmeans_model = KMeans(n_clusters=k, n_init='auto', random_state=15)
    kmeans_model.fit(data)
    sil_scores.append(silhouette_score(data, kmeans_model.labels_))

  #print(sil_scores)

  plt.plot(range(2, k_max+1), sil_scores)

  plt.title("Silhouette Scores vs K")
  plt.xlabel('Number of Clusters (K)')
  plt.ylabel('Score')

  plt.show()
```

```python
best_k = best_model_elbow(final_data, 9)
```
